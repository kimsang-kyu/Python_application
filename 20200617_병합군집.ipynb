## 병합 군집(Agglomerative clustering)

> 병합 군집 알고리즘
 
 - 각 포인트를 하나의 클러스터로 지정
 - 종료 조건을 만족할 때 까지 가장 비슷한 두 클러스터를 결정 

```
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```
###data load
import pandas as pd
iris = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/iris.csv')
```

```
###입력변수
x = iris.iloc[:,3:5]
y = iris.iloc[:,5]
```

```
### 병합 군진
agg = AgglomerativeClustering(n_clusters=3)
assigment = agg.fit_predict(x)
plt.scatter(x['Petal.Length'], x['Petal.Width'], c= assigment, s= 50)
```

```
from sklearn.preprocessing import StandardScaler
```

```
###표준화
x_std = StandardScaler().fit_transform(x)

### 병합군집
agg = AgglomerativeClustering(n_clusters=3)
assigment = agg.fit_predict(x_std)
plt.scatter(x_std[:,0], x_std[:,1], c= assigment, s= 50)
```

## 덴드로그램

```
from scipy.cluster.hierarchy import dendrogram, ward
```

```
#거리 정보가 담긴 배열
linkage_array = ward(x)
```

```
#덴드로그램
den = dendrogram(linkage_array)
```
